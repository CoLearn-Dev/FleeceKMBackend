```python
import pandas as pd
from lib.data_utils import fetch_and_prepare_data_all

DATASET_NAME = 'askme-10k'

try:
    data_df = pd.read_csv(f'data_samples/{DATASET_NAME}.csv')
    print('Dataset already exists.')
except FileNotFoundError:
    data_df = await fetch_and_prepare_data_all()
    data_df.to_csv(f'data_samples/{DATASET_NAME}.csv', index=False)
data_df
```

```python
sa_instance_states_columns = [col for col in data_df.columns if '_sa_instance_state' in col]
data_df = data_df.drop(columns=sa_instance_states_columns)
data_df
```

```python
data_df.columns
```

```python
df_sampled_1_question_per_paragraph = data_df.groupby('paragraph_id').apply(lambda x: x.sample(1)).reset_index(drop=True)
df_sampled_1_question_per_paragraph.to_csv(f'data_samples/{DATASET_NAME}-1-q-per-p.csv', index=False)
df_sampled_1_question_per_paragraph
```

```python
import numpy as np
from tqdm.notebook import tqdm
paragraph_ids = data_df['paragraph_id'].unique()
np.random.shuffle(paragraph_ids)

selected_paragraphs = []
total_rows = 0

for pid in tqdm(paragraph_ids):
    if total_rows >= 10000:
        break
    paragraph_rows = data_df[data_df['paragraph_id'] == pid]
    selected_paragraphs.append(paragraph_rows)
    total_rows += len(paragraph_rows)

df_sampled_all_questions_per_paragraph = pd.concat(selected_paragraphs).reset_index(drop=True)
df_sampled_all_questions_per_paragraph.to_csv(f'data_samples/{DATASET_NAME}-all-q-per-p.csv', index=False)
df_sampled_all_questions_per_paragraph
```
